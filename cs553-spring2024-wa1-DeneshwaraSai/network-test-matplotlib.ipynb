```python
import matplotlib.pyplot as plt

def readTestLatencyFromFile(filePath) :
    # These lists will be used to store the DNS names and average RTT (Round Trip Time) values read from a file.
    dnsNameList = list()
    avgRTTList = list()

    # It is used open a file in `filePath` with read mode. This allows the code to read the contents of the file.
    file = open(filePath, 'r')
    
    # The code block is iterating through each line in the file and splitting the line into a list of strings at every single space  method.
    for info in file:
        splitList = info.split()
        # 0th index = DNS name and 1st index = avg RTT value
        if(len(splitList) == 2):
            dnsNameList.append(splitList[0])
            avgRTTList.append(float(splitList[1]))

    print(dnsNameList)
    print(avgRTTList)

    # Rreturning a tuple containing the `dnsNameList` and `avgRTTList` lists which contains String of dns names and numeric of avg RTT value.
    return (dnsNameList, avgRTTList)

# The `constructPlotGraph` function is responsible for creating a line plot graph using the `matplotlib.pyplot` library.
# The graph has Domain names in X-axis.
# The graph has Avg RTT in Y-axis.
# The (15, 6) is the size of width and height of the graph.
# The Domain names are tilted to 90 degrees to show the string properly.
# tight_layout() method is used to design the graph with any overlapping. 
def constructPlotGraph(dnsNameList, avgRTTList):
    plt.figure(figsize=(15, 6))
    plt.xlabel("Domain names")
    plt.ylabel("Avg RTT")
    plt.plot(dnsNameList, avgRTTList)
    plt.title("Domain Names vs Average RTT")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()

'''
    The `constructBarGraph` function is responsible for creating a bar graph using the `matplotlib.pyplot` library.
    The graph has Domain names in X-axis.
    The graph has Avg RTT in Y-axis.
    The (15, 6) is the size of width and height of the graph.
    The Domain names are tilted to 45 degrees to show the string properly.
    tight_layout() method is used to design the graph with any overlapping 
'''
def constructBarGraph(dnsNameList, avgRTTList):
    plt.figure(figsize=(15, 6))
    plt.xlabel("Domain names")
    plt.ylabel("Avg RTT")
    plt.title("Domain Names vs Average RTT")
    plt.bar(dnsNameList, avgRTTList, color='blue')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()


# The code is reading data from a file called "network-test-latency.txt" using the `readTestLatencyFromFile` method and passing it as a method parameter. 
dnsNameList, avgRTTList = readTestLatencyFromFile("network-test-latency.txt")

# This function reads the DNS names and average RTT values from the file and returns them as two separate lists, `dnsNameList` and `avgRTTList`.
constructPlotGraph(dnsNameList, avgRTTList)

# This function reads the DNS names and average RTT values from the file and returns them as two separate lists, `dnsNameList` and `avgRTTList`.
constructBarGraph(dnsNameList, avgRTTList)

```
